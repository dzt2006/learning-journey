## 第四题

```python
score = ['Python成绩', 'SQL成绩']
```

```python
df['平均分'] = df[score].mean(axis=1).round(2)
df['总分'] = df[score].sum(axis=1)
df
```

```python
df.loc[df['总分'] == df['总分'].max(),:]
```

```python
df.loc[df['平均分'] >= 85,:].sort_values(by='平均分',ascending=False)
```

```python
def rank(a):
    if a >= 90:
        return 'A'
    elif a >= 80:
        return 'B'
    else:
        return 'C'
```

```python
df['python等级'] = df['Python成绩'].apply(rank)
df.loc[:,['姓名','Python成绩','python等级']].sort_values(by='python等级',ascending=False)
```

## 第五题

```python
import numpy as np
dates = pd.date_range("2026-05-01","2026-05-10")
products = ["商品A","商品B","商品C","商品D"]
np.random.seed(42)
```

```python
data = []
for date in dates:
    for product in products:
        sales = np.random.randint(1,1000)
        price = np.round(np.random.uniform(10,100),2)
        data.append([date,product,sales,price])

df = pd.DataFrame(data, columns=['日期','商品','销量','价格'])
df
```

```python
df.isna().sum()
```

```python
df['销售额'] = (df['销量'] * df['价格']).round(2)
df
```

```python
df.loc[df['销量'] == df['销量'].max(),:]
```

```python
dsales = df.groupby('日期')['销售额'].sum().reset_index()
days = dsales[dsales['销售额'] > 5000]['日期']
df.loc[df['日期'].isin(days), ['日期','销售额']]
```

```python
df.groupby('商品').agg(
    总销量=('销量','sum'),
    平均销量=('销量','mean'),
    最高销量=('销量','max'),
    总销售额=('销售额','sum'),
    平均销售额=('销售额','mean')
).round(2)
```
