第四题

```python
# 成绩数组 `scores = np.array([[85,92,78,90],[76,88,95,82],[93,80,87,91]])`
scores = np.array([[85,92,78,90],
                   [76,88,95,82],
                   [93,80,87,91]])
```

```python
# 计算每门课平均分、最高分、最低分
smean = scores.mean(axis=1)
smax = scores.max(axis=1)       # axis=0 是竖着算

smin = scores.min(axis=1)
print(f"每门课平均分：{smean}  最高分：{smax}  最低分：{smin}")
```

```python
# 计算每个学生总分与平均分
ssum = scores.sum(axis=1)
print(f"每个学生总分：{ssum}  平均分：{smean}")
```

```python
# 找出所有>90分的位置
print(np.where(scores > 90))    # 可以类比坐标 （1，2），（2，2），（0，1），（3，0）
```

```python
# 计算所有成绩的标准差与方差
STD = np.std(scores)
VAR = np.var(scores)
print(f"所有成绩标准差：{STD}  方差：{VAR}")
```

第五题

```python
np.random.seed(2026)
height = np.random.normal(165,8,50).round(1)
weight = np.random.normal(55,10,50).round(1)
```

```python
BMI = weight / (height/100) ** 2
BMI = np.round(BMI, 2)
print(BMI)
```

```python
# 按标准分类：偏瘦/正常/超重/肥胖
a = BMI < 18               # 偏瘦
b = (BMI>=18) & (BMI<24) # 正常
c = (BMI>=24) & (BMI<28)   # 超重
d = BMI >=28                # 肥胖
```

```python
# 统计各类别人数与占比
total = 50
c1 = np.sum(a)
c2 = np.sum(b)
c3 = np.sum(c)
c4 = np.sum(d)

# 统计占比
r1 = np.round(c1 / total * 100, 2)
r2 = np.round(c2 / total *100, 2)
r3 = np.round(c3 / total *100, 2)
r4 = np.round(c4 / total *100, 2)
```

```python
print("偏瘦", c1, f"{r1}%")
print("正常", c2, f"{r2}%")
print("超重", c3, f"{r3}%")
print("肥胖", c4, f"{r4}%")
```

```python
# 输出身高前5名的BMI数据
top_index = np.argsort(height)[::-1][:5]
print(f"身高前5位的位置：{top_index}")
top_BMI = BMI[top_index]
print(f"身高前5名的BMI数据：{top_index}")
```

```python
# 实现二维数组按行/按列分别求和、求均值
arr = np.random.randint(60, 101, size=(2, 2))
print(arr)

csum = arr.sum(axis=0)
cmean = arr.mean(axis=0)

rsum = arr.sum(axis=1)
rmean = arr.mean(axis=1)

print("按列求和：", csum)
print("按列均值：", cmean)
print("按行求和：", rsum)
print("按行均值：", rmean)
```
