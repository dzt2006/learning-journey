```python
import pandas as pd
import numpy as np
```

```python
# series 的创建 (列表)
s = pd.Series([10,2,3,4,5])
print(s)
```

```python
# 自定义索引
s = pd.Series([10,2,3,4,5], index=['a','b','c','d','e'])
print(s)
```

```python
# 定义name
s = pd.Series([10,2,3,4,5], index=['a','b','c','d','e'], name='月')
print(s)
```

```python
# 通过字典来创建
s = pd.Series({
    "a":1,
    "b":8,
    "c":4,
})
print(s)
```

```python
s1 = pd.Series(s,index=["a","c"])
print(s1)
```

```python
print(s)
print(s.index,s.values)
s.name = 'test'
print(s.dtype,s.name)
print(s.shape,s.ndim,s.size)
```

```python
# 显式索引
print(s.loc['a':'b'])
# 不支持切片
print(s.at['a'])
```

```python
# 隐式索引
print(s.iloc[2])
# 不支持切片
print(s.iat[2])
```

```python
# 访问数据
print(s['a'])
# 直接索引会报提醒
# print(s[1])
```

```python
print(s[s<5])
```

```python
s = pd.Series([1,2,3,4,5,6,7,8])
# 默认打印前 5 个
print(s.head())
```

```python
# 默认打印后 5 个
print(s.tail())
```

```python
# 常用函数
s = pd.Series([1,np.nan,None,9,6,10],index=['A','B','C','D','E','F'] ,name="test")
print(s)
```

```python
s.head(3)
```

```python
s.tail(2)
```

```python
# 查看所有的描述性信息  缺失值不统计
s.describe()
```

```python
# 获取元素个数
print(s.count())
```

```python
# 获取索引
print(s.keys()) # 方法
print(s.index)  # 属性
```

```python
# 检查 Series 里的每一个元素是否为缺失值
s.isna()
```

```python
# 用来检查每个元素是否在参数集合中
s.isin([12,4,9,10])
```

```python
print(s.mean(),s.max(),s.min())
print(s.sum(),s.median())   # 先忽略缺失值
print(s.std(),s.var())
```

```python
# 百分位数  计算值 9 + （10 - 9） * 0.25 = 9.25
s.quantile(0.75)
```

```python
# 众数
s['G'] = 10
print(s.mode())
```

```python
# 每个元素的计数 缺失值忽略
print(s.value_counts())
```

```python
# 按 索引 排序
s.sort_values()
```

```python
# 去重
s.drop_duplicates()
```

```python
# 去重
s.unique()
```

```python
print(s.nunique())
```

```python
# 按 值 排序
s.sort_index()
```
