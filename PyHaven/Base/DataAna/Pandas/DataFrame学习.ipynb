```python
import pandas as pd
```

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,10])

pd.DataFrame({
    "第一列":s1,
    "第二列":s2,
})
```

```python
# 通过字典创建
df = pd.DataFrame({
    "name":['tom','tom','sse','srr','dd','re'],
    "age":[15,15,18,18,19,20],
    "score":[52.3,52.3,60.2,85,42.5,55]
},index=[1,2,3,4,5,6],columns=["name","score","age"])
df
```

```python
# dataframe 的属性
print("行索引：")
print(df.index)
print("列标签：")
print(df.columns)
print("值")
print(df.values)
```

```python
print("维度：", df.ndim)
print("形状:",df.shape)
print("元素个数:",df.size)
print("数据类型：",df.dtypes)
```

```python
# 行列转置
print(df.T)
print(df.T.index)
```
